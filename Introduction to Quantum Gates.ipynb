```python
%matplotlib inline
# Importing standard Qiskit libraries and configuring account
from qiskit import QuantumCircuit, execute, Aer, IBMQ
from qiskit.compiler import transpile, assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
# Loading your IBM Q account(s)
provider = IBMQ.load_account()
```

```python
from qiskit.visualization import plot_bloch_multivector
```

```python
#Applying an X gate to a single qubit. Can do same for hadamard gate for 
circuit = QuantumCircuit(1, 1)
circuit.x(0)
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend = simulator ).result()
statevector = result.get_statevector()
statevector
```

```python
%matplotlib inline
#mpl stands for matplotlib
circuit.draw(output = 'mpl')
```

```python
#This is the favourite method of people to talk about quantum gates and statevectors
plot_bloch_multivector(statevector)
```

```python
#Run the circuit to do measurement
#The follwoing code measures qubit 0 and puts it in classical bit 0
circuit.measure([0], [0])
backend = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = backend, shots = 1024).result()
#Take the counts from the result
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
#To get matrix representation of the quantum gate
circuit = QuantumCircuit(1, 1)
circuit.x(0)
simulator = Aer.get_backend('unitary_simulator')
result = execute(circuit, backend = simulator ).result()
unitary = result.get_unitary()
unitary
```
